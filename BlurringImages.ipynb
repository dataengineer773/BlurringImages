```python
# Load libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load image as grayscale
image = cv2.imread("PLANEE.JPEG", cv2.IMREAD_GRAYSCALE)
```

```python
# Blur image
image_blurry = cv2.blur(image, (5,5))
```

```python
# Show image
plt.imshow(image_blurry, cmap="gray"), plt.axis("off")
plt.show()
```
